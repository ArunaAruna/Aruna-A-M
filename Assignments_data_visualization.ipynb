```python
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
a=pd.read_csv('https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/master/Iris.csv')
df=pd.DataFrame(a)
df=df.drop(['Id','Species'],axis=1)
#df.set_index('SepalLengthCm',inplace=True)
df
```

```python
fig = plt.figure(figsize=(10,10))

ax = plt.axes(projection='3d')
ax.scatter(df['SepalLengthCm'],df['SepalWidthCm'],marker='o',s=150)
ax.scatter(df['SepalLengthCm'],df['PetalLengthCm'],marker='o',s=100)
ax.scatter(df['SepalLengthCm'],df['PetalWidthCm'],marker='o',s=100)
plt.show()
```

1. Some values in the FlightNumber column are missing. These number are meant to increase by 10 with each row so 10055 and 10075 need to be put in place. Fill in these missing numbers and make the column an integer column(instead of a float column).

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({'From-To':['LoNDon_paris','MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis','Brussels_londOn'],
                     'FlightNumber':[10045,np.nan,10065,np.nan,10085],
                     'RecentDelays':[[23,47],[],[24,43,87],[13],[67,32]],
                     'Airline':['KLM(!)','<Air France>(12)','(British Airways.)','12.Air France',"Swiss Air"]})
df
  
```

```python
df['FlightNumber'].fillna(0,inplace=True)
df
```

```python
for i  in range(0,5):
    if df.loc[i,'FlightNumber']==0:
        df.loc[i,'FlightNumber']=df.loc[i-1,'FlightNumber']+10.0
    else:
        df.loc[i,'FlightNumber']=df.loc[i,'FlightNumber']
df
```

```python
df['FlightNumber']=df['FlightNumber'].astype('int32')

df
```

```python
df
```

2. The From_To column would be better as two separate column split each string on the underscore delimeter to give a new temporary DataFrame with the correct values. Assingn the correct column names to this temporary DataFrame.

```python
df_temp=df['From-To']
df_temp
```

```python
df_temp=df_temp.str.split('_',expand=True)
df_temp=df_temp.rename(columns={0:'From',1:'To'})
df_temp
```

3. Notice how the capitalization of the city names is all mixed up in this temporary DataFrme. Standardize the strings so that only the first letter is uppercase (e.g. "londON" should become "London")

```python
df_temp['From']=df_temp['From'].str.title()
df_temp['To']=df_temp['To'].str.title()
df_temp
```

4. Delete the From-To column from df and attach the temporary DataFrame from the previous question

```python
df_new=df.drop(columns='From-To')
df_new
```

```python
df_temp
```

```python
#concatenating both the data frames
df=pd.concat([df_temp,df_new],axis=1) 
df
```

5. In the RecentDelays column, the values have been entered into the DataFrame as a list. We would like each first value in its own column, each second  value in its own column, and so on. If there isn't an Nth value, the value should be NaN.<br> Expand the series of lists into a DataFrame named delays, rename the column delay_1, delay_2,etc. and replace the unwanted RecentDelays columns in df with delays.

```python
delays_new=df['RecentDelays']
delays_new
```

```python
delays=pd.DataFrame(delays_new)
delays
```

```python
for i in range(0,5):
    if len(delays['RecentDelays'][i])<3:
        for j in range(0,3-len(delays['RecentDelays'][i])):
            delays['RecentDelays'][i].append('NaN')

delays
```

```python
for i in range(0,5):
    delays.loc[i,'delay_1']=delays['RecentDelays'][i][0]
    delays.loc[i,'delay_2']=delays['RecentDelays'][i][1]
    delays.loc[i,'delay_3']=delays['RecentDelays'][i][2]
delays
```

```python
delays.drop(columns=['RecentDelays'],inplace=True)
delays
```

```python
df=pd.concat([df,delays],axis=1) 
df
```
